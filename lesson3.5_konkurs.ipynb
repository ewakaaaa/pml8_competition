```python
import pandas as pd
import numpy as np
np.random.seed(0)

from sklearn.metrics import mean_absolute_error

from sklearn.dummy import DummyRegressor
```

```python
train = pd.read_hdf('data/property.train.h5')
test = pd.read_hdf('data/property.test.h5') #zbi√≥r testowy bez odpowiedzi
```

```python
train.head()
```

```python
train['breadcrumbs_str'] = [' '.join([str(elem) for elem in s])  for s in train['breadcrumbs'] ]
train['breadcrumbs_num'] = pd.factorize( train['breadcrumbs_str'] )[0]
```

```python
train
```

```python
train[['breadcrumbs','breadcrumbs_str','breadcrumbs_num']]
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

test = train['breadcrumbs']

mlb = MultiLabelBinarizer()

res = pd.DataFrame(mlb.fit_transform(test),
                   columns=mlb.classes_,
                   index=test.index)
```

```python
res['–ú–æ—Å–∫–≤–∞']
```

## Basic Model

```python
feats = ['price']  ##cokolwiek na poczƒÖtek
X = train[ feats ].values
y = train["price"].values

model = DummyRegressor()
model.fit(X, y)
y_pred = model.predict(test[ "id" ].values)
```

## Submission
Trzeba przygotowaƒá plik, kt√≥ry bƒôdzie zawiera≈Ç: `id` i `price`.

```python
test['price'] = y_pred
test[ ['id', 'price'] ].to_csv('../output/model_dummy.csv', index=False) 
```
